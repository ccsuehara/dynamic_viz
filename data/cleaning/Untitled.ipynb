```python
import pandas as pd
```

```python
data = '../raw/BL2013_MF2599_v2.2.csv'
df = pd.read_csv(data)
```

```python
df.head()
```

```python
df = df[df['year']==2010]
```

```python
df.columns
```

```python
cols = ['country', 'year', 'lu', 'WBcode', 'region_code']
df = df[cols].reset_index(drop=True)
```

```python
rename_dic = {'lu': 'no_education',
              'WBcode': 'code'}
```

```python
df = df.rename(columns=rename_dic)
```

```python
export = '../clean/try.csv'
df.to_csv(export, index=False)
```
